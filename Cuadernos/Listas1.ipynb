# Listas

## Introducción

Las listas en python son colecciones mutables y ordenadas de datos. Los datos puedes ser de cualquier tipo (cadenas, números, objetos, etc).

Las listas pueden ser instanciadas con el constructor de la clase lista o utilizando el operador **[]**

```python
L1 = list() # Instancia una lista vacía
print( 'L1 =', L1 )
print( 'type( L1 ) =', type( L1 ) )
print( 'len( L1 ) =', len( L1 ) )
print()

L2 = [] # Instancia una lista vacía
print( 'L2 =', L2 )
print( 'type( L2 ) =', type( L2 ) )
print( 'len( L2 ) =', len( L2 ) )
print()
```

En los ejemplos anteriores, se dice que ambas listas carecen de dato o que son listas vacías.

Si se desea, pueden instanciarse listas inicializadas directamente:



```python
L3 = list( [0, 1, 2, 3, 4] ) # Instancia una lista con elementos
print( 'L3 =', L3 )
print( 'type( L3 ) =', type( L3 ) )
print( 'len( L3 ) =', len( L3 ) )
print()

L4 = [0, 1, 2, 3, 4] # Instancia una lista con elementos
print( 'L4 =', L4 )
print( 'type( L4 ) =', type( L4 ) )
print( 'len( L4 ) =', len( L4 ) )
print()
```

Las listas son llamadas mutables porque pueden ser modificadas.
Es decir, puede: agreagarse más elementos:

```python
L5 = [20, 30, 40]
print( 'L5 =', L5 )
print( 'type( L5 ) =', type( L5 ) )
print( 'len( L5 ) =', len( L5 ) )
print()

L5.append( 50 ) # Se agrega un elemento al final de la lista
print('Se agrega un elemento al final de la lista')
print( 'L5 =', L5 )
print( 'type( L5 ) =', type( L5 ) )
print( 'len( L5 ) =', len( L5 ) )
print()

L5.insert( 0, 10 ) # Se agrega un elemento al inicio de la lista
print('Se agrega un elemento al inicio de la lista')
print( 'L5 =', L5 )
print( 'type( L5 ) =', type( L5 ) )
print( 'len( L5 ) =', len( L5 ) )
print()
```

puede eliminarse elementos

```python
L6 = [10, 20, 30, 40, 50]
print( 'L6 =', L6 )
print( 'type( L6 ) =', type( L6 ) )
print( 'len( L6 ) =', len( L6 ) )
print()

L6.pop( ) # Se elimina el último elemento de la lista
print('Se elimina el último elemento de la lista')
print( 'L6 =', L6 )
print( 'type( L6 ) =', type( L6 ) )
print( 'len( L6 ) =', len( L6 ) )
print()

L6.pop( 0 )  # Se elimina el primer elemento de la lista
print('Se elimina el primer elemento de la lista')
print( 'L6 =', L6 )
print( 'type( L6 ) =', type( L6 ) )
print( 'len( L6 ) =', len( L6 ) )
print()

L6.remove( 30 )  # Se elimina el valor 30
print('Se elimina el valor 30')
print( 'L6 =', L6 )
print( 'type( L6 ) =', type( L6 ) )
print( 'len( L6 ) =', len( L6 ) )
print()
```

o puede modificarse un elemento en la lista:

```python
L7 = [10, 200, 30, 400, 50]
print( 'L7 =', L7 )
print( 'type( L7 ) =', type( L7 ) )
print( 'len( L7 ) =', len( L7 ) )
print()

L7[ 4 ] = 500 # Se modifica el valor en la última posición
print('Se modifica el valor en la última posición')
print( 'L7 =', L7 )
print( 'type( L7 ) =', type( L7 ) )
print( 'len( L7 ) =', len( L7 ) )
print()

L7[ 0 ] = 100  # Se modifica el valor en la primera posición
print('Se modifica el valor en la primera posición')
print( 'L7 =', L7 )
print( 'type( L7 ) =', type( L7 ) )
print( 'len( L7 ) =', len( L7 ) )
print()

L7[ 2 ] = 300  # Se modifica el valor en la posición 2
print('Se modifica el valor en la posición 2')
print( 'L7 =', L7 )
print( 'type( L7 ) =', type( L7 ) )
print( 'len( L7 ) =', len( L7 ) )
print()
```

Como puede observarse de los últimos ejemplos, el operador **[]** es utilizado para referirse al número de posición que ocupa un elemento en la lista. Es por esto que son llamadas ordenadas.

Es importante entender que las posiciones de los elementos están secuencialmente enumerados desde **cero** y **de uno en uno**.

Los datos almacenados en una lista **no** necesariamente están ordenados:

```python
L8 = [26, 16, 53, 40, 5, 84, 83, 26, 57, 81]
print( 'L8 =', L8 )
print( 'type( L8 ) =', type( L8 ) )
print( 'len( L8 ) =', len( L8 ) , '\n')
print( 'L8[ 0 ] =', L8[ 0 ])
print( 'L8[ 1 ] =', L8[ 1 ])
print( 'L8[ 2 ] =', L8[ 2 ])
print( 'L8[ 3 ] =', L8[ 3 ])
print( 'L8[ 4 ] =', L8[ 4 ])
print( 'L8[ 5 ] =', L8[ 5 ])
print( 'L8[ 6 ] =', L8[ 6 ])
print( 'L8[ 7 ] =', L8[ 7 ])
print( 'L8[ 8 ] =', L8[ 8 ])
print( 'L8[ 9 ] =', L8[ 9 ])
```

Dado que las listas son secuencias de elementos, el uso de ciclos es muy común:

```python
L9 = [26, 16, 53, 40, 5, 84, 83, 26, 57, 81]
print( 'L9 =', L9 )
print( 'type( L9 ) =', type( L9 ) )
print( 'len( L9 ) =', len( L9 ) , '\n')

i = 0
while i<len( L9 ):
    print( 'L9[',i,'] =', L9[ i ])
    i += 1
print()

for i in range(len( L9 )):
    print( 'L9[',i,'] =', L9[ i ])
```

## Inicialización

Dada la mutabilidad de las listas, puede agragarse repedidamente, con un ciclo, un dato a la vez.

Por ejemplo, puede agregarse la cantidad ( **n** ) de ceros que el usuario requiera:

```python
L = []
n = int(input('¿Cuántos ceros desea agregar a la lista? '))

for i in range(n):
  #print( 'L =', L , 'tamaño =', len(L)) # Esta línea es sólo demostrativa
  L.append( 0 )
print()

print('La lista queda:')
print( 'L =', L , 'tamaño =', len(L))
```

En la porción de código anterior, con cada ciclo que se agrega a la lista un **cero** (puede quitarse el **#** inicial de la línea 5 para corroborarlo).

### Lista de valores Enteros Aleatorios

La función **randrange** del módulo **random** retorna valores enteros aleatorios.

A continuación, se muestra cómo inicializar una lista con valores enteros aleatorios en el segmento **[ 0, superior )**

```python
from random import randrange

L = []
n = int(input('¿Cuántos ceros desea agregar a la lista? '))
superior = int(input('¿Cuál es el límite superior del segmento? '))

for i in range(n):
  #print( 'L =', L , 'tamaño =', len(L)) # Esta línea es sólo demostrativa
  L.append( randrange(superior) )
print()

print('La lista queda:')
print( 'L =', L , 'tamaño =', len(L))
```

A continuación, se muestra cómo inicializar una lista con valores enteros aleatorios en el segmento **[ inferior, superior )**

```python
from random import randrange

L = []
n = int(input('¿Cuántos ceros desea agregar a la lista? '))
inferior = int(input('¿Cuál es el límite inferior del segmento? '))
superior = int(input('¿Cuál es el límite superior del segmento? '))

for i in range(n):
  #print( 'L =', L , 'tamaño =', len(L)) # Esta línea es sólo demostrativa
  L.append( randrange(inferior,superior) )
print()

print('La lista queda:')
print( 'L =', L , 'tamaño =', len(L))
```

### Lista de valores ingresados por el usuario

Si se usa la función **input**, puede inicializarse una lista con **cadenas**:

```python
L = []
n = int(input('¿Cuántos ceros desea agregar a la lista? '))

for i in range(n):
  #print( 'L =', L , 'tamaño =', len(L)) # Esta línea es sólo demostrativa
  L.append( input( f'Ingresa L[{i}] ' ) )
print()

print('La lista queda:')
print( 'L =', L , 'tamaño =', len(L))
```

Si se requiere que sean **enteros**:

```python
L = []
n = int(input('¿Cuántos ceros desea agregar a la lista? '))

for i in range(n):
  #print( 'L =', L , 'tamaño =', len(L)) # Esta línea es sólo demostrativa
  L.append( int( input( f'Ingresa L[{i}] ' ) ) )
print()

print('La lista queda:')
print( 'L =', L , 'tamaño =', len(L))
```

Si se requiere que sean **flotantes**:

```python
L = []
n = int(input('¿Cuántos ceros desea agregar a la lista? '))

for i in range(n):
  #print( 'L =', L , 'tamaño =', len(L)) # Esta línea es sólo demostrativa
  L.append( float( input( f'Ingresa L[{i}] ' ) ) )
print()

print('La lista queda:')
print( 'L =', L , 'tamaño =', len(L))
```
