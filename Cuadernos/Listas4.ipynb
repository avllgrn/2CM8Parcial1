```python
def generaMatrizCeros(filas, columnas):
  if filas > 0 and columnas > 0:
    M = []  # Matriz (Lista de listas)
    for i in range(filas):
      F = [] #Fila (Lista de datos)
      for j in range(columnas):
        F.append( 0 )
      M.append(F)
    return M
  else:
    return M[[]]

def muestraMatriz(M):
  filas = len(M)
  columnas = len(M[0])
  for i in range( filas ):
    for j in range( columnas ):
        print( M[i][ j ], end='\t')
    print()
```

### Recorrido de una matriz (lista de listas)

Las matrices pueden ser recorridas con la pareja de ciclos que se ha visto en los ejemplos anteriores. Este recorrido se realiza:
1.    Fijo en una fila, se recorre de izquierda a derecha
2.    Habiendo terminado de recorrer la fila, se pasa hacia abajo a la siguiente

Por lo anterior, se dice que este recorrido es **de izquierda a derecha, y luego de arriba hacia abajo.**

```python
filas = 3
columnas = 4
M = generaMatrizCeros(filas, columnas)
print( 'M' )
muestraMatriz(M)
print()

contador = 1
# Se recorre la matriz, de izquierda a derecha de arriba a abajo
for i in range( filas ):
  for j in range( columnas ):
      M[i][ j ] = contador
      contador += 1   # Se rellena la matriz, con un contador para comprobar el recorrido

print( 'M' )
muestraMatriz(M)
```

Con sólo invertir el ciclo que controla los contadores, se rprovoca que el recorrido sea **de arriba hacia abajo, y luego de izquierda a derecha.**

La razón es que en un ciclo dentro de otro ciclo, el ciclo externo permanece en su iteración hasta que el interno termine todas las suyas.

```python
filas = 3
columnas = 4
M = generaMatrizCeros(filas, columnas)
print( 'M' )
muestraMatriz(M)
print()

contador = 1
# Se recorre la matriz, de izquierda a derecha de arriba a abajo
for j in range( columnas ):
  for i in range( filas ):
      M[i][ j ] = contador
      contador += 1   # Se rellena la matriz, con un contador para comprobar el recorrido

print( 'M' )
muestraMatriz(M)
```

### Buscar un dato en la matriz (lista de listas)

Se recorre la matriz (lista de listas), posición por posición, preguntando SI el dato buscado está en la posición actual; de ser localizado, se detiene la búsqueda; de no ser localizado, se contiúa la búsqueda. Si se recorrió toda la matriz (lista de listas) y nunca se localizó el dato buscado, el ciclo habrá completado todas sus vueltas.

```python
M = [[93, 39, 59, 66, 62], [30, 70, 0, 64, 44], [10, 29, 9, 5, 71]]
print( 'M' )
muestraMatriz(M)
print()

filas = len( M )
columnas = len( M[0] )
dato = int(input('Ingrese el dato a buscar: '))
print()

encontrado = False
# Se recorre la matriz, de izquierda a derecha de arriba a abajo
for i in range( filas ):
  for j in range( columnas ):
    if M[i][j] == dato:
      encontrado = True # Sirve para avisarle al ciclo externo que se detenga
      break # Sirve para detener el ciclo interno

  if encontrado:  # Sirve para detener el ciclo externo
      break

if (i<filas-1 and j<columnas-1) or dato==M[i][j]:
  print('El dato', dato, 'se encuentra en la posición [', i,'][', j, ']')
else:
  print('El dato NO se encuentra en M')
```

Gracias a las funciones de usuario, el código de la búsqueda puede minimizarse:

```python
def buscaEn(M, dato):
  filas = len( M )
  columnas = len( M[0] )

  # Se recorre la matriz, de izquierda a derecha de arriba a abajo
  for i in range( filas ):
    for j in range( columnas ):
      if M[i][j] == dato: # Si el dato es localizado, se retorna su posición, en forma de dupla
        return (i,j)

  # Si no es localizado, tras las vueltas de for, python retorna None


M = [[93, 39, 59, 66, 62], [30, 70, 0, 64, 44], [10, 29, 9, 5, 71]]
print( 'M' )
muestraMatriz(M)
print()

filas = len( M )
columnas = len( M[0] )
dato = int(input('Ingrese el dato a buscar: '))
print()

posicionDeDato = buscaEn(M, dato)
if isinstance(posicionDeDato, tuple): # Si la función retorna una dupla, el dato FUE localizado
  print('El dato', dato, 'se encuentra en la posición', posicionDeDato)

else:                                 # Si no retorna ninguna dupla, el dato NO fue localizado
  print('El dato no se encuentra en la lista')
```

### Contar ocurrencias de un dato en una matriz (lista de listas)

Se recorre la matriz (lista de listas), posición por posición, preguntando SI el dato buscado está en la posición actual; de ser localizado, se contabiliza la búsqueda; de no ser localizado, se contiúa la búsqueda. Tras recorrer toda la matriz (lista de listas), se retorna la cantidad de ocurrencias.

Sin función de usuario:

```python
M = [[93, 39, 59, 93, 62], [30, 70, 0, 93, 44], [93, 29, 9, 5, 71]]
print( 'M' )
muestraMatriz(M)
print()

filas = len( M )
columnas = len( M[0] )
dato = int(input('Ingrese el dato a buscar: '))
print()

ocurrencias = 0 # Inicialmente no se sabe cuántas ocurrencias hay, ni si las haya
# Se recorre la matriz, de izquierda a derecha de arriba a abajo
for i in range( filas ):
  for j in range( columnas ):
    if M[i][j] == dato:
      ocurrencias += 1

print('El dato', dato, 'se encuentra', ocurrencias, 'veces en la lista' )
```

Utilizando una función de usuario:

```python
def cuentaEn(M, dato):
  ocurrencias = 0 # Inicialmente no se sabe cuántas ocurrencias hay, ni si las haya
  # Se recorre la matriz, de izquierda a derecha de arriba a abajo
  for i in range( filas ):
    for j in range( columnas ):
      if M[i][j] == dato:
        ocurrencias += 1

  return ocurrencias    # Tras terminar el recorrido por toda la lista, se retorna la cantidad de ocurrencias

M = [[93, 39, 59, 93, 62], [30, 70, 0, 93, 44], [93, 29, 9, 5, 71]]
print( 'M' )
muestraMatriz(M)
print()

filas = len( M )
columnas = len( M[0] )
dato = int(input('Ingrese el dato a contar: '))
print()
ocurrencias = cuentaEn(M, dato)

print('El dato', dato, 'se encuentra', ocurrencias, 'veces en la lista' )
```

### Sumar matrices (listas de listas)

Sumar dos matrices (listas de listas) se realiza recorriéndola posición por posición, sumando elemento a elemento, sin una función de usuario:

```python
M1 = [[8, 5, 4, 5], [0, 7, 3, 10], [6, 8, 7, 0]]
M2 = [[7, 3, 3, 7], [3, 4, 6, 10], [2, 3, 9, 3]]

print( 'M1', M1 )
muestraMatriz(M1)
print()
print( 'M2', M2 )
muestraMatriz(M2)
print()

filas1 = len(M1)
columnas1 = len(M1[0])
filas2 = len(M2)
columnas2 = len(M2[0])

if filas1==filas2 and columnas1==columnas2:
    filas3 = filas1
    columnas3 = columnas2
    M3 = generaMatrizCeros(filas3, columnas3) # Se genera una matriz resultante, con las dimensiones de las matrices a sumar
    # print( 'M3', M3 )
    # muestraMatriz(M3)
    # print()

    for i in range( filas3 ):
        for j in range( columnas3 ):
            M3[i][j] = M1[i][j] + M2[i][j]  # Se suman elemento a elemento, de izquierda a derecha, de arriba hacia abajo

print( 'M3', M3 )
muestraMatriz(M3)
print()
```

Utilizando una función de usuario:

```python
def suma(M1, M2):
    filas1 = len(M1)
    columnas1 = len(M1[0])
    filas2 = len(M2)
    columnas2 = len(M2[0])

    if filas1==filas2 and columnas1==columnas2: #Si las matrices tienen las mismas dimensiones, se suman
        filas3 = filas1
        columnas3 = columnas2
        M3 = generaMatrizCeros(filas3, columnas3) # Se genera una matriz resultante, con las dimensiones de las matrices a sumar
        # print( 'M3', M3 )
        # muestraMatriz(M3)
        # print()

        for i in range( filas3 ):
            for j in range( columnas3 ):
                M3[i][j] = M1[i][j] + M2[i][j]  # Se suman elemento a elemento, de izquierda a derecha, de arriba hacia abajo

        return M3   # Se retorna la matriz resultante

    # En caso de ser de distintas dimensiones, Python retorna None

M1 = [[8, 5, 4, 5], [0, 7, 3, 10], [6, 8, 7, 0]]
M2 = [[7, 3, 3, 7], [3, 4, 6, 10], [2, 3, 9, 3]]
M3 = suma(M1, M2)

print( 'M1', M1 )
muestraMatriz(M1)
print()

print( 'M2', M2 )
muestraMatriz(M2)
print()

print( 'M3', M3 )
muestraMatriz(M3)
print()
```

### Multiplicar matrices (listas de listas)

Para multiplicar dos matrices (listas de listas), se multiplica cada fila de la primera por cada columna de la segunda. Por lo tanto, la restricción NO es que tengan las mismas dimensiones, sino que el número de columnas de la primera DEBE SER IDÉNTICO al número de filas de la seguna; es decir:

C <sub>eme x ele</sub> = A <sub>eme x ele</sub> *  B <sub>eme x ele</sub>

Y resulta una matriz (lista de listas) con el número de filas de la primera y el número de columnas de la segundo. Sin una función de usuario:

```python
M1 = [[4, 3, 5], [3, 5, 4]]
M2 = [[4, 1, 3, 5], [1, 2, 4, 0], [1, 1, 1, 4]]

print( 'M1', M1 )
muestraMatriz(M1)
print()
print( 'M2', M2 )
muestraMatriz(M2)
print()

filas1 = len(M1)
columnas1 = len(M1[0])
filas2 = len(M2)
columnas2 = len(M2[0])

if columnas1==filas2:   # Si el número de columnas de la primera es idéntico al de la segunda, se multiplican
    filas3 = filas1
    columnas3 = columnas2
    M3 = generaMatrizCeros(filas3, columnas3) # Se genera una matriz resultante, con las dimensiones de las matrices a multiplicar
    # print( 'M3', M3 )
    # muestraMatriz(M3)
    # print()

    for i in range( filas1 ):
        for j in range( columnas2 ):
            for k in range( columnas1 ):
                M3[i][j] += M1[i][k] * M2[k][j]  # Se multiplican cada fila de la primera matriz, por cada columna de la segunda

print( 'M3', M3 )
muestraMatriz(M3)
print()
```

Utilizando una función de usuario:

```python
def multiplica(M1, M2):
    filas1 = len(M1)
    columnas1 = len(M1[0])
    filas2 = len(M2)
    columnas2 = len(M2[0])

    if columnas1==filas2:   # Si el número de columnas de la primera es idéntico al de la segunda, se multiplican
        filas3 = filas1
        columnas3 = columnas2
        M3 = generaMatrizCeros(filas3, columnas3) # Se genera una matriz resultante, con las dimensiones de las matrices a multiplicar
        # print( 'M3', M3 )
        # muestraMatriz(M3)
        # print()

        for i in range( filas1 ):
            for j in range( columnas2 ):
                for k in range( columnas1 ):
                    M3[i][j] += M1[i][k] * M2[k][j]  # Se multiplican cada fila de la primera matriz, por cada columna de la segunda

        return M3   # Se retorna la matriz resultante

    # En caso de ser de distintas dimensiones, Python retorna None

M1 = [[4, 3, 5], [3, 5, 4]]
M2 = [[4, 1, 3, 5], [1, 2, 4, 0], [1, 1, 1, 4]]
M3 = multiplica(M1, M2)

print( 'M1', M1 )
muestraMatriz(M1)
print()

print( 'M2', M2 )
muestraMatriz(M2)
print()

print( 'M3', M3 )
muestraMatriz(M3)
print()
```

### Buscar el menor dato de una matriz (lista de listas)

Se supone que el dato en la primer fila y primer columna es el menor, se compara éste contra cada dato en la matriz (lista de listas). Sin una función de usuario:

1.    Se supone que el dato en la primer fila y primer columna es el menor
2.    Para cada posición en la matriz (lista de listas)
      1.    Se compara el menor actual contra el dato en la posición actual
      2.    Si se encuentra un dato menor, se almacena el nuevo
      3.    Si no, se pasa a la siguiente posición
3. Al final del recorrido, se tiene el  menor de los datos en la matriz (lista de listas)

Sin una función de usuario:

```python
M = [[61, 41, 31, 96, 47], [13, 61, 2, 24, 6], [39, 46, 72, 11, 30], [80, 85, 22, 34, 90], [66, 64, 24, 84, 30], [26, 89, 31, 60, 91]]
filas = len(M)
columnas = len(M[0])

menor = M[0][0] # Se supone que el dato en la primera fila y primera columna es el menor
filaMenor = 0
columnaMenor = 0
# Se recorre la matriz, de izquierda a derecha de arriba a abajo
for i in range( filas ):
    for j in range( columnas ):
        if M[i][j] < menor: # Si se encuentra un dato menor, se almacena el nuevo
            menor = M[i][j]
            filaMenor = i
            columnaMenor = j
# Al final del recorrido, se tiene el menor de los datos en la matriz

print( 'M', M )
muestraMatriz(M)
print()

print(f'El menor es {menor}, que se encuentra en [{filaMenor}][{columnaMenor}]')
```

Utilizando una función de usuario:

```python
def buscaMenorEn(M):
    filas = len(M)
    columnas = len(M[0])

    menor = M[0][0] # Se supone que el dato en la primera fila y primera columna es el menor
    filaMenor = 0
    columnaMenor = 0
    # Se recorre la matriz, de izquierda a derecha de arriba a abajo
    for i in range( filas ):
        for j in range( columnas ):
            if M[i][j] < menor: # Si se encuentra un dato menor, se almacena el nuevo
                menor = M[i][j]
                filaMenor = i
                columnaMenor = j
    # Al final del recorrido, se tiene el menor de los datos en la matriz

    return (filaMenor, columnaMenor)

M = [[61, 41, 31, 96, 47], [13, 61, 2, 24, 6], [39, 46, 72, 11, 30], [80, 85, 22, 34, 90], [66, 64, 24, 84, 30], [26, 89, 31, 60, 91]]
print( 'M', M )
muestraMatriz(M)
print()
i, j = buscaMenorEn(M)
print(f'El menor es {M[i][j]}, que se encuentra en [{i}][{j}]')
```

### Copiar una matriz (lista de listas)

Es muy importante recordar que las listas son referenciadas por un nombre; por lo que escribir **lo siguiente**, **NO ES COPIARLA** sino ponerle dos nombres (referencias) a la misma lista:

```python
M1 = [[1, 4, 2, 3], [2, 3, 1, 2], [1, 1, 2, 1]]
M2 = M1                                         # Esto NO es copiar una lista
print( 'M1', M1 )
print( 'M2', M2 )
print()

# Si M2 fuera copia de M1, al modificar M1 NO debería ser modificada M2
M1[0][0] = 100
# sin embargo, el cambio se ve en "las dos" porque NO hay dos,
# sino que hay UNA SOLA lista.
# M1 y M2  son DOS NOMBRES (referencias) para LA MISMA lista:

print( 'M1', M1 )
print( 'M2', M2 )
print()
```

**Copiar una lista** es agregar una copia de cada dato, posición por posición, de una lista a otra.

Sin una función de usuario:

```python
M1 = [[1, 4, 2, 3], [2, 3, 1, 2], [1, 1, 2, 1]]
filas1 = len(M1)
columnas1 = len(M1[0])
M2 = generaMatrizCeros(filas1, columnas1)
filas2 = len(M2)
columnas2 = len(M2[0])

print( 'M1', M1 )
print( 'M2', M2 )
print()

# Se recorre la matriz, de izquierda a derecha de arriba a abajo
for i in range( filas2 ):
    for j in range( columnas2 ):
        M2[i][j] = M1[i][j] # Copiando cada valor de M1 en M2

print( 'M1', M1 )
print( 'M2', M2 )
print()

# Si M2 fuera copia de M1, al modificar M1 NO debería ser modificada M2
M1[0][0] = 100
print( 'M1', M1 )
print( 'M2', M2 )
print()
```

Utilizando una función de usuario:

```python
def generaCopiaDe(M1):
    filas1 = len(M1)
    columnas1 = len(M1[0])
    M2 = generaMatrizCeros(filas1, columnas1)
    filas2 = len(M2)
    columnas2 = len(M2[0])

    print( 'M1', M1 )
    print( 'M2', M2 )
    print()

    # Se recorre la matriz, de izquierda a derecha de arriba a abajo
    for i in range( filas2 ):
        for j in range( columnas2 ):
            M2[i][j] = M1[i][j] # Copiando cada valor de M1 en M2

    print( 'M1', M1 )
    print( 'M2', M2 )
    print()

    return M2

M1 = [[1, 4, 2, 3], [2, 3, 1, 2], [1, 1, 2, 1]]
M2 = generaCopiaDe(M1)

# Si M2 fuera copia de M1, al modificar M1 NO debería ser modificada M2
M1[0][0] = 100
print( 'M1', M1 )
print( 'M2', M2 )
print()
```
